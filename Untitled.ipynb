```python
import matplotlib.pyplot as pt
import pandas as pd
data = pd.read_csv('energydata_trainning.csv')
appliences_energy = data[['Appliances','WeekStatus']]
appliences_energy
```

```python
appliences_energy_weekday = appliences_energy[data.WeekStatus == 'Weekday']
appliences_energy_weekday
```

```python
avg_energy_weekday = appliences_energy_weekday['Appliances'].mean()
avg_energy_weekday
```

```python
appliences_energy_weekend = appliences_energy[data.WeekStatus == 'Weekend']
appliences_energy_weekend
```

```python
avg_energy_weekend = appliences_energy_weekend['Appliances'].mean()
avg_energy_weekend
```

```python
appliance_energy = [avg_energy_weekday,avg_energy_weekend]
week_days = ['weekday','weekend']
pt.bar(week_days,appliance_energy,color='g',width=0.30,label='Appliences')
pt.xlabel('Weekdays')
pt.ylabel('Appliences energy in weekday')
pt.title('Average applienes energy in weekday')
pt.legend(loc=(1.1,0.4))
```
