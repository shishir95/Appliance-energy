```python
import matplotlib.pyplot as pt
import pandas as pd


data = pd.read_csv('energydata_trainning.csv')
temp_humidity = data[['T8','RH_8','Day_of_week']]
temp_humidity
```

```python
temp_humidity_monday = temp_humidity[data.Day_of_week == 'Monday']
temp_humidity_monday
```

```python
avg_temp_monday = temp_humidity_monday['T8'].mean()
avg_temp_monday
```

```python
temp_humidity_tuesday = temp_humidity[data.Day_of_week == 'Tuesday']
temp_humidity_tuesday
```

```python
avg_temp_tuesday = temp_humidity_tuesday['T8'].mean()
avg_temp_tuesday
```

```python
temp_humidity_wednesday = temp_humidity[data.Day_of_week == 'Wednesday']
temp_humidity_wednesday
```

```python
avg_temp_wednesday = temp_humidity_wednesday['T8'].mean()
avg_temp_wednesday
```

```python
temp_humidity_thursday = temp_humidity[data.Day_of_week == 'Thursday']
temp_humidity_thursday
```

```python
avg_temp_thursday = temp_humidity_thursday['T8'].mean()
avg_temp_thursday
```

```python
temp_humidity_friday = temp_humidity[data.Day_of_week == 'Friday']
temp_humidity_friday
```

```python
avg_temp_friday = temp_humidity_friday['T8'].mean()
avg_temp_friday
```

```python
temp_weekday = [avg_temp_monday,avg_temp_tuesday,avg_temp_wednesday,avg_temp_thursday,avg_temp_friday]
week_days = ['monday','tuesday','wednesday','thursday','friday']
pt.bar(week_days,temp_weekday,color='g',width=0.50,label='temperature')
pt.xlabel('Weekdays')
pt.ylabel('Tempertures')
pt.title('Average  teenger room temperatures in weekday')
pt.legend(loc=(1.1,0.4))
```

```python
humidity_weekday = [avg_humidity_monday,avg_humidity_tuesday,avg_humidity_wednesday,avg_humidity_thursday,avg_humidity_friday]
week_days = ['monday','tuesday','wednesday','thursday','friday']
pt.bar(week_days,humidity_weekday,color='b',width=0.50,label='Humidity')
pt.xlabel('Weekdays')
pt.ylabel('Humidity')
pt.title('Average teenger humidity in weekday')
pt.legend(loc=(1.1,0.4))
```
