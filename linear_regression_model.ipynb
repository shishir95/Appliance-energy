```python
import matplotlib.pyplot as pt
import pandas as pd
from mpl_toolkits.mplot3d import axes3d
from sklearn import linear_model



data = pd.read_csv('energydata_trainning.csv')
data
```

```python
energy_data = data[['T1','RH_1','T2','RH_2','T3','RH_3','T4','RH_4','T5','RH_5','T6','RH_6','T7','RH_7','T8','RH_8','T9','RH_9']]
energy_data
```

```python
tmp=data.loc[:, ['T1', 'T2','T3','T4','T5','T6','T7','T8','T9']]
tmp
```

```python
x1=tmp.mean(axis=1)
x1
```

```python
h=data.loc[:, ['RH_1', 'RH_2','RH_3','RH_4','RH_5','RH_6','RH_7','RH_8','RH_9']]
x2=h.mean(axis=1)
x2
```

```python
data_frame = pd.DataFrame({
    'Appliances': data.Appliances,
    'X1': x1,
    'X2': x2
}
)
data_frame
```

```python
linear_regression = linear_model.LinearRegression()
linear_regression.fit(data_frame[['X1','X2']],data_frame.Appliances)
```

```python
linear_regression.coef_
```

```python
linear_regression.intercept_
```
