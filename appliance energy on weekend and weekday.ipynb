```python
import matplotlib.pyplot as pt
import pandas as pd


data = pd.read_csv('energydata_trainning.csv')
appliences_energy = data[['Appliances','Day_of_week']]
appliences_energy 
```

```python
appliences_energy_monday = appliences_energy[data.Day_of_week == 'Monday']
appliences_energy_monday
```

```python
avg_energy_monday = appliences_energy_monday['Appliances'].mean()
avg_energy_monday
```

```python
appliences_energy_tuesday = appliences_energy[data.Day_of_week == 'Tuesday']
appliences_energy_tuesday
```

```python
avg_energy_tuesday = appliences_energy_monday['Appliances'].mean()
avg_energy_tuesday
```

```python
appliences_energy_wednesday = appliences_energy[data.Day_of_week == 'Wednesday']
appliences_energy_wednesday
```

```python
avg_energy_wednesday = appliences_energy_monday['Appliances'].mean()
avg_energy_wednesday
```

```python
appliences_energy_thursday = appliences_energy[data.Day_of_week == 'Thursday']
appliences_energy_thursday
```

```python
avg_energy_thursday = appliences_energy_monday['Appliances'].mean()
avg_energy_thursday
```

```python
appliences_energy_friday = appliences_energy[data.Day_of_week == 'Friday']
appliences_energy_friday
```

```python
avg_energy_friday = appliences_energy_monday['Appliances'].mean()
avg_energy_friday
```

```python
appliences_energy_saturday = appliences_energy[data.Day_of_week == 'Saturday']
appliences_energy_saturday
```

```python
avg_energy_saturday = appliences_energy_monday['Appliances'].mean()
avg_energy_saturday
```

```python
appliences_energy_sunday = appliences_energy[data.Day_of_week == 'Sunday']
appliences_energy_sunday
```

```python
avg_energy_sunday = appliences_energy_monday['Appliances'].mean()
avg_energy_sunday
```

```python
energy_weekday = [avg_energy_monday,avg_energy_tuesday,avg_energy_wednesday,avg_energy_thursday,avg_energy_friday]
week_days = ['monday','tuesday','wednesday','thursday','friday']
pt.bar(week_days,energy_weekday,color='g',width=0.50,label='Appliences')
pt.xlabel('Weekdays')
pt.ylabel('Appliences energy in weekday')
pt.title('Average applienes energy in weekday')
pt.legend(loc=(1.1,0.4))
```

```python
energy_weekend = [avg_energy_saturday,avg_energy_sunday]
week_days = ['saturday','sunday']
pt.bar(week_days,energy_weekend,color='b',width=0.50,label='Humidity')
pt.xlabel('Weekends')
pt.ylabel('Appliences energy in weekend')
pt.title('Average appliences in weekend')
pt.legend(loc=(1.1,0.4))
```
